# 변수 (Variable)
데이터를 저장하는 **메모리 공간**의 이름

## 변수 사용 이유
1. 유지보수
2. 재사용
3. 가독성

```python
# '=' 연산자: 오른쪽 값을 왼쪽 변수에 대입(할당) //대입 연산자//
team_name = "kjy"
print(team_name)

# 값 변경 : 기존 데이터 위에 새로운 값이 씌워짐 (재할당)
team_name = 'kkjj'
print(team_name)

# 값 복사 : 변수에 담긴 값을 다른 변수에 할당 가능 
my_team = team_name
print(my_team)
```

## 변수 이름 짓기(Nameing rule)
### 1) 필수 규칙 (위반하면 에러)
- 사용 가능 문자: 영문 대소문자, 숫자, 언더바
- 숫자로 시작 불가
- 특수문자/공백 불가
- 예약어 사용 불가
### 2) 권장 관례 (가독성을 위한 약속)
- snake_case : 단어 사이를 언더바로 연결
- 직관적 명령
- 한글 사용 가능하지만 인코딩/호환성 이슈 있을 수 있으므로 지양

```python
# 대소문자 구문
X = 1
x = 2
print(X,x)

# 숫자 사용
user1 = 'zz'
user_2 = 'ㅇㅇ'
print(user1, user_2)

# 스네이크 케이스
teamname = 'kfffj'
team_name ='권장사항'

# 한글 값 사용
팀명 = 'ㅏㅇㅇ'
print(팀명)

# 에러 케이스
# 1st_user ='숫자 시작 안됨'
# user-name = '하이픈 안됨'
# user email = '공백 안됨'
# class = '예약어 안됨' 

# 예약어 목록 확인
import keyword
print('파이썬 예약어 목록')
print(keyword.kwlist)

# 의미 있는 이름 사용
a = True
is_user_logged_in = True
```
