# 반복문 for
- 리스트/문자열 등의 요소를 하나씩 꺼내서 처리하거나 정해진 횟수만큼 반복
## for-in 문(Iterable 객체 순회)
- 문자열, 리스트, 튜플, 딕셔너리, 세트 등 순회 가능한 객체의 요서를 하나씩 꺼냄

```python
# 리스트 및 문자열 순회
ohgiraffers = ['pig', 'bear', 'goriaal']
for teacher in ohgiraffers: 
    print(f"{teacher} (이름 길이: {len(teacher)})")

for word in 'python':
    print(word, end= ' ')

# 딕셔너리 순회
user_info = {'bear':'곰', 'tiger':'호랑이', 'monkey':'원숭이'}
for key in user_info:
    print('key={key}, Value:{user_info[key]}')
```

## range()
- 숫자 범위를 생성하는 함수
- range(시작, 끝, 증감)의 형태로 작성하며 끝값은 포함되지 않음

```python
# 1부터 10까지의 합계
total = 0
for i in range(1, 11):
    total +=1
print(f"1부터 10까지 합계: {total}")
print(f"1부터 10까지의 합계: {sum(range(1, 11))}")

# 1 ~ 100 사이의 홀수의 합
n_sum = 0
for n in range(1, 101, 2):
    n_sum += n
print(f"1부터 100까지 홀수의 합계: {n_sum}")
print(f"1부터 100까지 홀수의 합계: {sum(range(1, 101, 2))}")
```

# 중첩 반복문 (Nested Loop)
- 반복문 안에 반복문이 있는 구조

```python
# 구구단
for i in range(2, 10):
    print(f"==={i}단===")
    for j in range(1, 10):
        print(f'{i}*{j}={i*j}') 
    print() # n단이 종료된 후 계행
```

## enumerate()와 zip()
- enumerate(): 순번(index)과 값을 동시에 꺼낼 때 사용
- zip() : 여러개의 리스트를 동시에 순회할 때 사용

```python
# enumerate(): 몇 번째 요소인지 인덱스가 필요할 때
fruits = ['사과', '포도', '체리']
for idx, fruit in enumerate(fruits, start=1):
    print(f'{idx}번째 과일 : {fruit}')

# zip
names = ['홍길동', '유관순', '신사임당']
scores = [90, 95, 87]
for name, score in zip(names, scores):
    print(f'{name}의 점수는 {score}점입니다.')
```
