# 반복문 while
- 특징: 조건문이 참(True)인 동안 실행 구문을 계속해서 반복

```python
# 기본 while 문 : 1~10까지 츨력
i = 1
while i <= 10:  #조건식(참인 동안 반복)
    print(i, end=' ')
    i+= 1       #증감식
```

## 루프의 흐름 제어 : break vs continue
- break : 반복문을 즉시 종료하고 탈출
- continue : 반복문의 남은 코드를 건너뛰고 다음 반복(조건 검사)으로 이동

```python
# break 예제 -> 5를 만나면 탈출하는 예제
i = 0
while i < 10:
    i += 1
    if i == 5:
        print('5 발견! 중단', end = ' ')
        break
    print(i, end=' ')

print('\n')

# continue 예제 -> 1~10까지 출력하되, 3의 배수는 건너뛰기
i = 0
while i < 10:
    i+= 1
    if i% 3 ==0:
        print('3의 배수 패스!', end = ' ')
        continue
    print(i, end=' ')   #continue를 만나게 되면 동작하지 않는 코드
```

## 무한 루프와 break
- 반복 횟수를 미리 알 수 없을 때 while True를 사용하고, 특정 조건에서 break로 탈출한다.

```python
menu = """
SK 구내식당 메뉴 
1. 된장찌개
2. 김치찌개
3. 주문종료
"""

order = []

while True:
    choice = input(menu)
    match choice:
        case "1":
            order.append('된장찌개')
            print('된장찌개가 추가되었습니다.')
        case "2":
            order.append('김치찌개')
            print('김치찌개가 추가되었습니다.')
        case "3":
            print('주문을 종료합니다.')
            break
        case _:
            print('잘못 입력하셨습니다. 다시 선택해주세요.')
print(f"최종 주문 내역: {order}")
```
