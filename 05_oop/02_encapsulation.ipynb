# 캡슐화
- private 속성(변수)을 선언해서 외부에서 직접 접근을 막고, 제공된 메소드를 통해서 사용하도록 강제한다.
- 주요한 데이터가 외부에 노출되지 않는 은닉화 효과를 얻을 수 있다.

```python
class Account: 
    def __init__(self, username, balance):
        # private 접근 제한 처리
        self.__username = username
        self.__balance = balance

    def deposit(self, money):
        '''
        입금 메소드
        - money 유효성 검사
        - 오염된 입금 처리 방지
        '''
        if money <= 0:
            print('입금 액이 유효하지 않습니다.')
        else:
            self.__balance += money
            print('입금이 정상처리 되었습니다.')

    def withdraw(self, money):
        if money <= 0:
            print('출금액이 유효하지 않습니다.')
        elif money > self.__balance:
            print('잔액이 부족합니다.')
        else:
            self.__balance -= money
            print('출금이 정상처리 되었습니다.')

    def info(self):
        print(f'{self.__username}님의 계좌의 현재 잔액은 {self.__balance}원 입니다.')

```

# 파이썬에서 private 는 '약속'이다
- _클래스명__변수명으로 이름이 바뀐 것이므로 접근하려고 하면 접근 가능하다.
- 하지만 개발자 사이의 약속으로 외부에서 접근하지 않도록 한다.

```python
acc = Account('hong', 1_000_000)
# print(acc.username)
# print(acc.balance)
# 권장하지 않는 방법
print(acc._Account__username)
print(acc._Account__balance)
```

## 왜 캡술화를 할까?
- `acc.balance -= 1억` 처럼 직접 값을 수정할 수 있다면 은행 프로그램은 망가지게 된다.
- deposit(), withdraw()와 같이 메소드라는 필터를 거쳐서 잘못된 데이터가 객체 안으로 들어오는 것을 막는 것이 캡슐화의 핵심이다.

```python
# 입금하기 
acc.deposit(500_000)
acc.deposit(-500_000)
acc.info()

# 출금하기
acc.withdraw(700_000)
acc.withdraw(-700_000)
acc.withdraw(40_000_000)
acc.info()
```
