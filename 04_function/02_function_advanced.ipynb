# 함수 심화

## 가변 인자 패킹 (*args, **kwargs)
- *args : 여러 개의 위치 인자를 튜플로 묶어서 받음
- **kwargs : 여러 개의 키워드 인자를 딕셔너리로 묶어서 받음

```python
def test_all(*args, **kwargs):
    print(f'args: {args}')
    print(f'kwargs: {kwargs}')

test_all(1)
test_all(1, 2)
test_all(1, 2, name='홍길동')
test_all(3, nmae='홍길동', age=20)
```

# 언패킹 (Unpacking)
- 리스트나 딕셔너리를 풀어서 함수의 인자로 전달할 때 사용한다.

```python
def sum_abc(a, b, c):
    return a + b + c

nums = [10, 20, 30]
print(sum_abc(*nums))       #리스트 언패킹

info = {'a': 1, 'b': 2, 'c': 3}
print(sum_abc(**info))       #딕셔너리 키워드 언패킹
```

# 데이터 전달 방식(Call by Object Reference)
- 불변 : int, str, tuple 등 -> 값 복사 처럼 동작한다. 원본은 유지된다.
- 가변 : list, dict, set 등 -> 참조(주소 값)가 공유된다. 원본이 수정된다.

```python
# 불변 테스트
def func1(a):
    a = 200         # 불변이므로 새로운 값이 할당되어 변화가 없다

num = 100
func1(num)
print(f'불변 원본(int): {num}')

# 가변 테스트
def func2(b):
    b[1] *= 100     # 가변이므로 원본 내부의 값이 직접 수정된다.

my_list = [1, 2, 3]
func2(my_list)
print(f'가변 원본(list): {my_list}')
```

# 타입 힌트(Type Hint)
- 동적 타이핑(Dynamic Typing) : 파이썬은 변수 자체에 자료형이 없고, 대입 되는 값에 따라 타입이 결정 된다.
- 타입 힌트란 정적 코드 분석 도구(ex. VS Code 등)가 타입을 체크할 수 있도록 돕는다.
- [중요] : 타입 힌트는 가이드일 뿐 실제 실행 시 파이썬 엔진에 의해 강제되지 않는다.

```python
# 동적 타이핑 확인
x = '안녕'
x = 123             # 다른 자료형을 제대입해도 오류 발생 X
print(type(x))

# 타입 힌트의 기본
greeting_msg: str = 'Hello'
greeting_msg = 123      # 힌트는 str이지만 int를 넣어도 에러 없이 정상 동작
print(f'greeting_msg: {greeting_msg}')
```

```python
# 다양한 자료형의 타입 힌트 작성법
# 변수명 뒤에 : 을 적어주며 복합 자료형은 내부 요소의 타입까지 명시한다.

# 기본 자료형
n : float = 123.456
score : int = 90

# 컬렉션 자료형
nums : list[int] = [1, 2, 3]
user: tuple[int, str, str] = (1, '홍길동', 'hong')

# | 기호는 OR(또는) 을 의미하는 union 타입이다
info: dict[str, str | int] = {'name' : '신사임당', 'age': 42}
```

## 함수에서의 타입 힌트 활용
- 매개변수에는 : 타입을, 반환 값에는 -> 타입을 사용하여 명시한다.

```python
def repeat_message(message: str, count: int) -> str:
    return message * count


result = repeat_message('파이썬', 3)
print(f'결과 : {result}')

result_error = repeat_message(10, '2')
print(f'에러 결과 : {result_error}')
```
