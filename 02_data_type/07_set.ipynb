# 집합 (Set)
- 순서가 없고, **중복을 허용하지 않는** 자료구조
- 수학의 집합 개념을 코드로 구현한 것으로, 중복 제거에 탁월하다.

````python
# 생성 및 중복 제거
safari_set = {"Bear", "Koala", "Gorilla", "Bear"}
print(f"집합출력: {safari_set}")

# 형변환을 이용한 중복 제거 (자주 쓰이는 패턴)
numbers_list = [1, 2, 3, 3, 3, 4, 4, 5]
unique_numbers = set(numbers_list)
print(f"리스트에서 중복 제거 : {unique_numbers}")
````

## [참고] 집합과 정렬(sorted)
- 집합은 순서가 없기 때문에 제거한 후 다시 순서대로 나열하고 싶다면 sorted() 함수를 사용한다.
- sorted()는 어떤 자료형이든 정렬되 '리스트' 로 반환한다.

````python
# 중복 제거 후 오름차순 정렬
fruits = ["사과", "바나나", "사과", "포도", "체리"]
unique_fruits = set(fruits)
sorted_fruits = sorted(unique_fruits)
print(f"정렬된 과일 목록: {sorted_fruits}")
````

## 집합의 특징
1. 인덱싱 불가 : 순서가 없어서 'set[0]' 처럼 접근하면 에러가 발생한다.
2. 가변성(Mutable) : 요소를 추가(add)하거나 제거(remove)할 수 있다.

````python
ohgiraffers = {"Sheep", "Squirrel", "Bear"}

ohgiraffers.add("Gorilla")
ohgiraffers.discard("Elephant") # 제거 : 해당 값이 없어도 오류 발생하지 않음
ohgiraffers.remove("Bear") # 제거 : 해당 값이 없으면 에러 발생

# ohgiraffrs[0] -> set은 인덱싱 불가

ohgiraffers
````

## 수학적 집합 연산
파이썬 집합은 기호(연산자)를 통해 직관적으로 연산할 수 있다.

````python
set_a = {'Bear','Squirrel', 'Sheep', 'Gorilla'}
set_b = {'Gorilla', 'Wolf', 'Tiger', 'Bear'}

# 합집합
print(f"합집합: {set_a | set_b}")
print(f"합집합: {set_a.union (set_b)}")

# 교집합
print(f"교집합: {set_a & set_b}")
print(f"교집합: {set_a.intersection (set_b)}")

# 차집합
print(f"차집합: {set_a - set_b}")
print(f"차집합: {set_a.difference (set_b)}")

# 대칭 차집합
print(f"대칭 차집합: {set_a ^ set_b}")
print(f"대칭 차집합: {set_a.symmetric_difference (set_b)}")
````

## 문제 2. 다음 코드 중 에러가 발생하는 행을 고르고, 그 이유를 설명
```python
my_set = {1, 2, 3}
my_list = [3, 4, 5]
print(my_set.intersection(my_list))  # 1번
print(my_set & my_list)              # 2번
```

### 정답 : 2번 -> 집합 연산자는 반드시 양 쪽 모두 집합 자료형이어야 함. 메소드는 다른 자료형을 형변환 후 연산함.

````python
my_set = {1, 2, 3}
my_list = [3, 4, 5]
print(my_set.intersection(my_list))  
print(my_set & my_list)  
````
