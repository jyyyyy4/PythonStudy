# None 타입
'값이 없음'을 명시적으로 나타내는 자료형

```python
# 선언: 아무런 값도 담고 있지 않은 상태
empty_value = None
print(empty_value, type(empty_value))

# 특징 : 논리형으로 판단 시 False와 유사하게 취급됨
print(bool(None))
```

# None을 사용하는 이유
-변수를 미리 만들어두고 싶지만, 아직 넣을 값이 없을 때
-함수의 결과 값이 없음을 나타낼 때

```python
user_name = None
# 나중에 값이 들어오면 업데이트
user_name='다람쥐'
print(user_name)
```
