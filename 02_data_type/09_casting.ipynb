# 형변환 (Casting)
서로 다른 자료형 간 연산 등을 위하여 다른 자료형으로 타입을 변환하는 것이다.

## 암묵적 형변환(Implicit)
파이썬이 연산 과정에서 데이터 손실을 최소화하기 위해 자동으로 타입을 맞추는 과정이다.

````python
print(1+ 2.3)   # 1.0 + 2.3 = 3.3 (int가 float로 변환)
print(True + 2) # 1 + 2 =3 (bool가 int로 변환)
print(False + 5.5)  # 0.0 + 5.5 = 5.5 (bool가 float로 변환)
````

## 명시적 형변환(Explicit)
개발자가 의도를 가지고 직접 함수를 호출하여 타입을 바꾸는 것이다.

### 숫자와 문자 간의 형변환
- int(value) : 정수로 변환(실수는 소수점 아래를 버림)
- float(value) : 실수로 변환
- str(value) : 어떤 자료형이든 문자열로 변환

````python
# [주의] 소수점이 포함된 문자열은 바로 int로 변환 불가
int("3")
# int("3.14")
int(float('3.14'))

# [참고] 실수를 정수로 변경 시
int(5.99)   # 반올림 X 소수점 버림 O
````

### 데이터 구조 간의 변화
데이터의 성격을 바꾸고 싶을 때 활용한다.

````python
# 튜흘 <-> 리스트 : 수정 가능 여부 변경
tp = (1, 2, 3)
lst = list(tp)  # 수정 가능한 리스트로 변환

# 리스트 <-> 집합 : 중복 제거
names = ['tiger', 'bear', 'tiger']
unique_names = set(names)   # 중복 제거

# 딕셔너리 -> 리스트 (기본적으로는 key 값만 추출)
user = {'id': 1, 'name': 'Sheep'}
print(list(user))
print(list(user.values()))
````

### 논리형으로의 형변환
조건문에서 가장 많이 쓰이는 개념. **값이 있으면 True, 비어있으면 False라고 기억** 한다.

````python
# 데이터가 있는 경우 -> True
print(f"숫자 100: {bool(100)}")
print(f"문자 ' ': {bool(' ')}")
print(f"리스트 [1, 2]: {bool([1, 2])}")

# 데이턱 비어 있는 경우 => False
print(f"숫자 0: {bool(0)}")
print(f"문자 '': {bool('')}")
print(f"리스트 []: {bool([])}")
````

````python
# 실전 활용 예시
# 리스트에 요소가 들어 있는지 확인할 때 len(lst) > 0 라고 작성할 필요가 없다.
#lst = [1, 2]
lst = []
if lst:
    print(f"리스트에 {len(lst)}개의 데이터가 존재합니다.")
else:
    print(f"리스트가 비어있습니다.")
````
